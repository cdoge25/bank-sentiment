```python
import pandas as pd
import google_play_scraper as app
from app_store_scraper import AppStore
```

```python
review_list = []

result, continuation_token = app.reviews(
    'vn.com.techcombank.bb.app',
    lang='vi', # defaults to 'en'
    country='vn', # defaults to 'us'
    sort=app.Sort.NEWEST, # defaults to Sort.NEWEST
)

for item in result:
    review_list.append(item)
```

```python
n=1
while True:
    start_count = len(review_list)
    result, continuation_token = app.reviews(
        'vn.com.techcombank.bb.app',
        continuation_token=continuation_token
    )
    for item in result:
        review_list.append(item)
    print(f"loop {n}, collected {len(review_list)} items total")
    n=n+1
    last_count = len(review_list)
    if last_count==start_count:
        break
```

```python
df = pd.DataFrame.from_records(review_list)
```
